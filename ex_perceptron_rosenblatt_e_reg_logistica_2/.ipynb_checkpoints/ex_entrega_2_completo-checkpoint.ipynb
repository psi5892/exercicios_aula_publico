```python
import torch
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv("circles_and_squares.csv")
data = data_df.to_numpy()
Xd = data[:800, :]
Xd_test = data[800:, :]
```

```python
def Rosenblatt_minibatch(Xd, eta, Nt, M, Nb, Ne):
    Nmb = int(np.floor(Nt / Nb))
    W = np.zeros((Ne * Nmb + 1, M))
    y_mb = np.zeros((Nmb, Nb))
    e_mb = np.zeros((Nmb, Nb))
    eta = eta / Nb
    J_MSE = np.zeros((Ne, 1))
    m = 0
    for k in range(Ne):
        np.random.shuffle(Xd)
        X = Xd[:, 0:M]
        d = Xd[:, M]
        for l in range(Nmb):
            d_mb = d[l * Nb : (l + 1) * Nb]
            X_mb = X[l * Nb : (l + 1) * Nb, :]
            y_mb[l, :] = np.sign(X_mb @ W[m, :])
            e_mb[l, :] = d_mb - y_mb[l, :]
            G = X_mb.T @ e_mb[l, :]
            W[m + 1, :] = W[m, :] + eta * G.T
            m = m + 1
            J_MSE[k] = J_MSE[k] + (np.linalg.norm(e_mb[l, :])) ** 2
        J_MSE[k] = J_MSE[k] / (Nmb * Nb)
    return y_mb, e_mb, W, J_MSE
```

```python
eta = 0.001  # passo de adaptação do algoritmo de Rosenblatt
M = 400  # número de pesos (Notar que não consideramos bias neste problema)
Nt = len(Xd)
Nb = 128
Ne = 3000
(y, e, W, J_MSE) = Rosenblatt_minibatch(Xd, eta, Nt, M, Nb, Ne)
```

```python
plt.figure()
# Mostra a função custo ao longo das épocas
plt.plot(J_MSE, ".-r")
plt.ylabel("J_MSE")
plt.grid(axis="x", color="0.5")
plt.grid(axis="y", color="0.5")
plt.xlabel("épocas")
plt.axis([0, Ne, min(J_MSE) - 0.01, max(J_MSE) + 0.01])
```

```python
X_test = Xd_test[:, 0:-1]  # sinal de entrada
d_test = Xd_test[:, -1].reshape(-1, 1)  # sinal desejado
W_last = W[int(Ne * np.floor(Nt / Nb)), :].reshape(-1, 1)  # W na última iteração
y_test = np.sign(X_test @ W_last)

print("Percentual de acertos (teste)")
sum(y_test == d_test)[0]/len(X_test)*100
```

```python
X_train = Xd[:, :-1]
d_train = Xd[:, -1].reshape(-1, 1)
y_train = np.sign(X_train @ W_last)

print("Percentual de acertos")
sum(y_train == d_train)[0]/len(X_train)*100
```
